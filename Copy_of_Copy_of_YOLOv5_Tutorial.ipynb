**Corrected code for YOLO result**

````
import torch
from PIL import Image

# Model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/best.pt')  # local model

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['image_name', 'xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'class', 'name'])

for im in int_values:
    # Inference
    image_path = '/content/drive/MyDrive/wet_video_to_frames/cam_angle_1_1/frame' + str(im) + '.jpg'
    results = model(image_path)

    # Extract the detections from the results
    detections = results.pandas().xyxy[0]

    # Remove the '.jpg' extension from the image name
    image_name = f'frame{im}'
    detections['image_name'] = image_name

    # Append the detections to the result_df
    result_df = result_df.append(detections)

result_df.reset_index(drop=True, inplace=True)
print(result_df)
````

**Video to Frames Extraction**

````
import cv2
````

````
cam =cv2.VideoCapture('/content/drive/MyDrive/GDToT/1688452096551444764_cam_1.mp4')
````

````
fps=cam.get(cv2.CAP_PROP_FPS)
fps
````

````
import os
import cv2

# Directory where you want to save the images
output_folder = '/content/drive/MyDrive/GDToT/cam-angle_1_1'

# Check if the output folder exists; if not, create it
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

n = 0
i = 0
while True:
    ret, frame = cam.read()
    if n % 3 == 0:  # Capture a frame every 3rd frame (10fps)
        # Construct the output image file path
        output_path = os.path.join(output_folder, '{}.jpg'.format(i))

        # Save the frame to the output path
        cv2.imwrite(output_path, frame)

        i += 1
    n += 1
    if ret is False:
        break

cam.release()
cv2.destroyAllWindows()
````
